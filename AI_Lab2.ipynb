```python
# Load the image
img = cv2.imread('wellfound.jpeg')

plt.imshow(img)
plt.title("Original (BGR) Image")
plt.show()
```

```python
# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis('off')
plt.show()
```

```python
# Resize to 128x128
img_resized = cv2.resize(img_rgb, (128, 128))

# Display resized image
plt.imshow(img_resized)
plt.title("Resized (128x128) Image")
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale
img_gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

# Display grayscale image
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Apply binary thresholding
_, img_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

# Display binary image
plt.imshow(img_binary, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
plt.show()
```

```python
# Normalize to range 0 to 1
img_normalized = img_resized / 255.0

# Display normalized image
plt.imshow(img_normalized)
plt.title("Normalized Image")
plt.axis('off')
plt.show()
```

```python
# Horizontal Flip
img_flipped = cv2.flip(img_resized, 1)

# Rotate 90 degrees clockwise
img_rotated = cv2.rotate(img_resized, cv2.ROTATE_90_CLOCKWISE)

# Display augmented images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img_flipped)
plt.title("Horizontally Flipped")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_rotated)
plt.title("Rotated 90°")
plt.axis('off')

plt.show()
```

```python
# Apply Gaussian blur
img_gaussian = cv2.GaussianBlur(img_resized, (5, 5), 0)

# Apply Median blur
img_median = cv2.medianBlur(img_resized, 5)

# Display blurred images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img_gaussian)
plt.title("Gaussian Blur")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_median)
plt.title("Median Blur")
plt.axis('off')

plt.show()
```

```python
# Flatten the grayscale image
flattened = img_gray.flatten()

# Show the shape
print("Flattened image shape:", flattened.shape)
print(flattened[:10])  # Show first 10 pixel values
```

### Summary of Image Preprocessing Steps:
- Loaded and displayed image
- Converted BGR to RGB
- Resized image to 128x128
- Converted to grayscale
- Applied binary thresholding
- Normalized pixel values
- Performed image augmentation (flip, rotate)
- Applied Gaussian and Median filters
- Flattened image for ML input
